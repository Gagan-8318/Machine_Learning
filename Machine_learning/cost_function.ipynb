```python
import numpy as np  
def cost(x,y,b0,b1):
    m=len(x)
    total_error=0
    for i in range(m):
        y_pred = b0 + b1 * x[i]
        total_error+=y[i]-(b0+b1*x[i])
    return total_error/m
  
```

```python
x = np.array([1,2,3,4,5])
y = np.array([200,40,600,80,100])

# Suppose coefficients
b0 = 0   # intercept
b1 = 20  # slope

print("Cost:", cost(x, y, b0, b1))
```
